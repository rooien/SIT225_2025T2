## Setup and Listener 

```python
import time, datetime
from collections import deque
import pandas as pd
import random   # replace this with your ArduinoCloud client

# Buffers
buffer_incoming = deque(maxlen=5000)  # continuous feed
buffer_plot = []                      # isolated batch for plotting
N = 1000                              # batch size for plotting

def get_new_sample():
    
    return {
        "t": datetime.datetime.now(),
        "x": random.uniform(-1,1),
        "y": random.uniform(-1,1),
        "z": random.uniform(-1,1)
    }

def listen_loop(iterations=3000, delay=0.01):
    for _ in range(iterations):
        sample = get_new_sample()
        buffer_incoming.append(sample)
        time.sleep(delay)

print("Listener ready. Call listen_loop() to simulate streaming data.")
```
